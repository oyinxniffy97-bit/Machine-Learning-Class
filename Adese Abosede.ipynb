```python
import matplotlib.pyplot as plt
import numpy as np

# Create spiral data
theta = np.linspace(0, 8*np.pi, 1000)
r = theta**2
x = r * np.cos(theta)
y = r * np.sin(theta)

# Create the plot
plt.figure(figsize=(10, 10))
plt.scatter(x, y, c=theta, cmap='twilight', s=10, alpha=0.8)
plt.title('Spiral Galaxy - Created by Abosede', fontsize=18, fontweight='bold')
plt.xlabel('X Dimension', fontsize=14)
plt.ylabel('Y Dimension', fontsize=14)
plt.colorbar(label='Spiral Intensity')
plt.grid(True, alpha=0.2)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure with subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Data Visualization Portfolio - By Oyinx', fontsize=20, fontweight='bold')

# Plot 1: Line chart
x1 = np.linspace(0, 10, 100)
axs[0, 0].plot(x1, np.sin(x1), 'b-', linewidth=2, label='Sine Wave')
axs[0, 0].plot(x1, np.cos(x1), 'r--', linewidth=2, label='Cosine Wave')
axs[0, 0].set_title('Trigonometric Functions', fontsize=14)
axs[0, 0].legend()
axs[0, 0].grid(True, alpha=0.3)

# Plot 2: Bar chart
categories = ['Python', 'Machine\nLearning', 'Data\nScience', 'Jupyter']
values = [95, 85, 90, 88]
colors_bar = ['#3776ab', '#ff6b6b', '#4ecdc4', '#f4a259']
axs[0, 1].bar(categories, values, color=colors_bar, alpha=0.8)
axs[0, 1].set_title('Skills Proficiency', fontsize=14)
axs[0, 1].set_ylabel('Percentage (%)')

# Plot 3: Scatter plot
x3 = np.random.randn(200)
y3 = np.random.randn(200)
colors_scatter = np.random.rand(200)
axs[1, 0].scatter(x3, y3, c=colors_scatter, cmap='viridis', s=50, alpha=0.7)
axs[1, 0].set_title('Random Distribution', fontsize=14)
axs[1, 0].grid(True, alpha=0.3)

# Plot 4: Pie chart
sizes = [30, 25, 20, 25]
labels = ['Analysis', 'Modeling', 'Visualization', 'Reporting']
colors_pie = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
axs[1, 1].pie(sizes, labels=labels, colors=colors_pie, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Project Time Distribution', fontsize=14)

plt.tight_layout()
plt.show()
```
